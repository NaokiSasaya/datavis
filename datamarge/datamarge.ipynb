```python
import csv

def create_country_iso3_mapping(csv_file):
    country_iso3_mapping = {}
    with open(csv_file, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            country_name = row['Country name']
            iso3_code = row['ISO3']
            country_iso3_mapping[country_name] = iso3_code
    return country_iso3_mapping

# CSVファイルから対応表を作成
csv_file = 'data/master2.csv'
country_iso3_mapping = create_country_iso3_mapping(csv_file)



        
```

```python
country_iso3_mapping
```

```python



# 対応表の使用例
country_name = 'Japan'
iso3_code = country_iso3_mapping.get(country_name)
print(iso3_code)  # 結果: AFG

        
```

```python
life['ISO3']=life['ISO3'].astype(object)
```

```python
for index,row in life.iterrows():
    country=row['Country name']
    iso3=country_iso3_mapping.get(country)
    print(country,iso3)
    life.loc[index,'ISO3']=iso3
```

```python
life.to_csv('life_iso3.csv')
```
